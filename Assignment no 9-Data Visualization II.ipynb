```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=sns.load_dataset('titanic')
```

```python
sns.boxplot(x='sex',y='age',hue='survived',data=dataset)
```

```python
sns.boxplot(x='survived',y='age',hue='sex',data=dataset)
```

```python
sns.scatterplot(x='survived',y='age',hue='sex',data=dataset)
```

```python
sns.swarmplot(x='survived',y='age',hue='sex',data=dataset,size=4)
```

```python
sns.stripplot(x='survived',y='age',hue='sex',data=dataset)
```
